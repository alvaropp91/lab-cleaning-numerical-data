```python
# Rename the columns so they follow the PE8 (snake case) (lower_case_with_underscores)
customer_df.columns = customer_df.columns.str.lower().str.replace(' ', '_')
```

```python
# Change effective to date column to datetime format.
customer_df["effective_to_date"] = pd.to_datetime(customer_df['effective_to_date'])
```

```python
# Create a NEW dataframe using only the numerical data called numerical_df. Use this for the following questions.
numericals_df = customer_df.select_dtypes(np.number)
numericals_df
```

```python
# Define a function that differentiates between continuous and discrete variables. Hint: The number of unique values might be useful. Store continuous data into a continuous_df variable and do the same for discrete_df

def differentiate_variables(numericals_df):
    continuous_variables = []
    discrete_variables = []

    for column in numericals_df.columns:
        variable = numericals_df[column]
        if isinstance(variable[0], (int, float)):
            continuous_variables.append(column)
        else:
            discrete_variables.append(column)

    continuous_df = numericals_df[continuous_variables]
    discrete_df = numericals_df[discrete_variables]

    return continuous_df, discrete_df
```

```python
continuous_df, discrete_df = differentiate_variables(numericals_df)
```

```python
# Create a function to plot every discrete variable. Do the same with the continuous variables (Be Careful, you may need to change the plot type to one better suited for continuous data!)

import seaborn as sns

def plot_discrete_variables(discrete_df):
    for col in discrete_df.columns:
        sns.displot(discrete_df[col], kde=False)
        
    return plot_discrete_variables

plot_discrete_variables(discrete_df)
```

```python
def plot_continuous_variables(continuous_df):
    for col in continuous_df.columns:
        sns.boxplot(data=continuous_df[col])
        plt.xlabel(col)
        plt.show()

plot_continuous_variables(continuous_df)
```

```python
# Check number of unique values in each column, can they be combined in any way to ease encoding? Comment your thoughts and make those changes.

def check_categorical_variables(categoricals_df):
    for col in categoricals_df.columns:
        unique_values = categoricals_df[col].unique()
        num_unique_values = len(unique_values)
        data_type = categoricals_df[col].dtype

        print(f"Column: {col}")
        print(f"Data Type: {data_type}")
        print(f"Unique Values: {unique_values}")
        print(f"Number of Unique Values: {num_unique_values}")
        print("----------------------")

check_categorical_variables(categoricals_df)
```

```python
# Group policy column to have less unique values:

categoricals_df["policy"] =np.where(categoricals_df["policy"].isin(["Corporate L1", "Corporate L2", "Corporate L3"]), "Corporate", categoricals_df["policy"])
categoricals_df["policy"] =np.where(categoricals_df["policy"].isin(["Personal L1", "Personal L2", "Personal L3"]), "Personal", categoricals_df["policy"])
categoricals_df["policy"] =np.where(categoricals_df["policy"].isin(["Special L1", "Special L2", "Special L3"]), "Special", categoricals_df["policy"])
```

```python
# Delete the column "policy_type" because it contains the same information as the "policy" column
categoricals_df = categoricals_df.drop(['policy_type'], axis=1)
```

```python
# Save the cleaned catagorical dataframe as categorical.csv You will use this file again this week.
categoricals_df.to_csv('categorical.csv', index=False)
```
